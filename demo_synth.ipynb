# Stimulus synthesis demo

This notebook allows you to explore the stimuli used in the study
'The consonance of the carillon' by Peter M. C. Harrison and James M. C. MacConnachie.
To generate a stimulus, run the cell below. You can change the `lower_pitch` and `interval` parameters to generate different stimuli. 

```python
import IPython
from synth import synth_stimulus

path = "demo_stimulus.wav"
lower_pitch = 66  # in the experiment this ranged from 65-67
interval = 7  # in the experiment this ranged from 0-15

synth_stimulus(
    path=path,
    lower_pitch=lower_pitch,
    upper_pitch=lower_pitch + interval,
)
IPython.display.Audio(path, autoplay=True)
```
